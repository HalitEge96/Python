```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from google.colab import files
```

```python
url = 'https://raw.githubusercontent.com/Ari-vu/SML/main/Given_data/train.csv'
df = pd.read_csv(url)
```

```python
X_train = df.drop(columns=['Lead'])
y_train = df.Lead
norm = MinMaxScaler().fit(X_train)
X_train_norm = norm.transform(X_train)
```

```python
model = LogisticRegression(solver='liblinear', random_state=43)
model.fit(X_train_norm, y_train)
```

```python
url = 'https://raw.githubusercontent.com/Ari-vu/SML/main/test.csv'
df2 = pd.read_csv(url)
```

```python
X_test_norm = norm.transform(df2)
```

```python
df2['Lead'] = model.predict(X_test_norm)
df2['prediction'] = np.where(df2['Lead']=='Female',1,0)
df2.head()
df2['prediction'].to_csv('output.csv', header=False, line_terminator = ',',
                         index=False, encoding ='utf-8-sig') 
files.download('output.csv')
```
